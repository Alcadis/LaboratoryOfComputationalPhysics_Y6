````python
#5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:
#where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.
#In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.

# Define the densities dictionary and radii list
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}#linear, superficiale and volumetric densities
radii = [1,2,3]# vari raggi 

# Define a list of lambda functions using comprehensions
# These lambda functions calculate the circumference, area, and volume for a given radius
functions = [lambda r: 2*3.14*r,
             lambda r: 3.14*r**2,
             lambda r: (4/3)*3.14*r**3
            ]
# Calculate and print the weight for each combination of radius and density using zip 
weights = [(material, r, [func(r) * density for func, density in zip(functions, density_values)])
           for material, density_values in densities.items()#prima condizione
           for r in radii] #seconda condizione

      
for material, r, weight in weights:
    print(f"For {material}, radius {r}:")#parte dal primo materiale primo raggio e va avanti fino all'ultimo materiale ultimo raggio
    print(f"Weight of the circle: {weight[0]:.2f}")#posizione 1 nella lista dei pesi, troncato al secondo decimale
    print(f"Weight of the disk: {weight[1]:.2f}")
    print(f"Weight of the sphere: {weight[2]:.2f}")
    print()







````

#```python
# Parent class
class Dog:

    # Class attribute
    species = 'mammal'

    # Initializer / Instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.is_hungry = True

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)
    
    # ADDED instance method
    def eat(self):
        self.is_hungry = False

# Child class (inherits from Dog class)
class RussellTerrier(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)

# Child class (inherits from Dog class)
class Bulldog(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)

# Function to check hunger status of all dogs
def check_hunger_status(dogs):
    if all(not dog.is_hungry for dog in dogs):
        print("My dogs are not hungry.")
    else:
        print("My dogs are hungry.")

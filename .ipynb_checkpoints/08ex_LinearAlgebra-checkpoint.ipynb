```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
#1
N = 1000
x1 = np.random.normal(0, 1, N)
x2 = x1 + np.random.normal(0, 3, N)
x3 = 2 * x1 + x2
data = np.array([x1, x2, x3]).T #trasposto
#2
cov_matrix = np.cov(data.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
#3
U, S, Vt = np.linalg.svd(data - np.mean(data, axis=0), full_matrices=False)
svd_eigenvalues = S**2 / (N - 1)
svd_eigenvectors = U
#4
total_var = np.sum(eigenvalues)
explained_var = [(i / total_var)*100 for i in sorted(eigenvalues, reverse=True)]
cumulative_var = np.cumsum(explained_var)
n_components = np.argmax(cumulative_var >= 99) + 1 #n components per avere 99% variability
#5
pca = PCA(n_components=n_components)
data_transformed = pca.fit_transform(data)
#6
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Original Data Plots
axs[0, 0].scatter(data[:, 0], data[:, 1], alpha=0.2)
axs[0, 0].set_title('Original Data (x1 vs x2)')
axs[0, 1].scatter(data[:, 0], data[:, 2], alpha=0.2)
axs[0, 1].set_title('Original Data (x1 vs x3)')
axs[0, 2].scatter(data[:, 1], data[:, 2], alpha=0.2)
axs[0, 2].set_title('Original Data (x2 vs x3)')

# Transformed Data Plots
axs[1, 0].scatter(data_transformed[:, 0], data_transformed[:, min(1, n_components - 1)], alpha=0.2)
axs[1, 0].set_title('PCA Transformed Data (PC1 vs PC2)' if n_components > 1 else 'PCA Transformed Data (PC1)')

if n_components > 1:
    axs[1, 1].scatter(data_transformed[:, 0], data_transformed[:, min(2, n_components - 1)], alpha=0.2)
    axs[1, 1].set_title('PCA Transformed Data (PC1 vs PC3)' if n_components > 2 else 'PCA Transformed Data (PC1 vs PC2)')
    axs[1, 2].scatter(data_transformed[:, min(1, n_components - 1)], data_transformed[:, min(2, n_components - 1)], alpha=0.2)
    axs[1, 2].set_title('PCA Transformed Data (PC2 vs PC3)' if n_components > 2 else 'PCA Transformed Data (PC1 vs PC2)')
else:
    # Clear plots if n_components is 1
    axs[1, 1].axis('off')
    axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

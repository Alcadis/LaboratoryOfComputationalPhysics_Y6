1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
#metodo pca
#1
N = 1000
x1 = np.random.normal(0, 1, N)
x2 = x1 + np.random.normal(0, 3, N)
x3 = 2 * x1 + x2
data = np.array([x1, x2, x3]).T
#2
cov_matrix = np.cov(data)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
#3
U, S, Vt = np.linalg.svd(data, full_matrices=False)
#4
variability_percentages = (eigenvalues / np.sum(eigenvalues)) * 100
scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)
pca = PCA(n_components=3)
pca.fit(data_standardized)
reduced_data = pca.transform(data_standardized)
#5
data_new = pca.inverse_transform(reduced_data)
data_new_transposed = data_new.T
#6
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
# Original Data plots
axs[0, 0].scatter(data[:,0], data[:,1], alpha=0.5)
axs[0, 0].set_xlabel('x1')
axs[0, 0].set_ylabel('x2')
axs[0, 0].set_title('Original Data (x1 vs x2)')

axs[0, 1].scatter(data[:,0], data[:,2], alpha=0.5)
axs[0, 1].set_xlabel('x1')
axs[0, 1].set_ylabel('x3')
axs[0, 1].set_title('Original Data (x1 vs x3)')

axs[0, 2].scatter(data[:,1], data[:,2], alpha=0.5)
axs[0, 2].set_xlabel('x2')
axs[0, 2].set_ylabel('x3')
axs[0, 2].set_title('Original Data (x2 vs x3)')

# Redefined Data plots
axs[1, 0].scatter(data_new_transposed[0,:], data_new_transposed[1,:], alpha=0.5)
axs[1, 0].set_xlabel('PC1')
axs[1, 0].set_ylabel('PC2')
axs[1, 0].set_title('Redefined Data (PC1 vs PC2)')

axs[1, 1].scatter(data_new_transposed[0,:], data_new_transposed[2,:], alpha=0.5)
axs[1, 1].set_xlabel('PC1')
axs[1, 1].set_ylabel('PC3')
axs[1, 1].set_title('Redefined Data (PC1 vs PC3)')

axs[1, 2].scatter(data_new_transposed[1,:], data_new_transposed[2,:], alpha=0.5)
axs[1, 2].set_xlabel('PC2')
axs[1, 2].set_ylabel('PC3')
axs[1, 2].set_title('Redefined Data (PC2 vs PC3)')

plt.show()
```

```python
#metodo svd
#1
N = 1000
x1 = np.random.normal(0, 1, N)
x2 = x1 + np.random.normal(0, 3, N)
x3 = 2 * x1 + x2
data = np.array([x1, x2, x3])
#2
cov_matrix = np.cov(data)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
#3
U, S, Vt = np.linalg.svd(data, full_matrices=False)
#4
variability_percentages = (eigenvalues / np.sum(eigenvalues)) * 100
S[-1] = 0
data_reduced = np.diag(S)
#5
data_new = U @ data_reduced @ Vt
data_new_transposed = data_new.T
#6
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
# Original Data plots
axs[0, 0].scatter(data[0, :], data[1, :], alpha=0.5)
axs[0, 0].set_xlabel('x1')
axs[0, 0].set_ylabel('x2')
axs[0, 0].set_title('Original Data (x1 vs x2)')

axs[0, 1].scatter(data[0, :], data[2, :], alpha=0.5)
axs[0, 1].set_xlabel('x1')
axs[0, 1].set_ylabel('x3')
axs[0, 1].set_title('Original Data (x1 vs x3)')

axs[0, 2].scatter(data[1, :], data[2, :], alpha=0.5)
axs[0, 2].set_xlabel('x2')
axs[0, 2].set_ylabel('x3')
axs[0, 2].set_title('Original Data (x2 vs x3)')
#Redefined Data plots
axs[1, 0].scatter(data_new_transposed[:, 0], data_new_transposed[:, 1], alpha=0.5)
axs[1, 0].set_xlabel('PC1')
axs[1, 0].set_ylabel('PC2')
axs[1, 0].set_title('Redefined Data (PC1 vs PC2)')

axs[1, 1].scatter(data_new_transposed[:, 0], data_new_transposed[:, 2], alpha=0.5)
axs[1, 1].set_xlabel('PC1')
axs[1, 1].set_ylabel('PC3')
axs[1, 1].set_title('Redefined Data (PC1 vs PC3)')

axs[1, 2].scatter(data_new_transposed[:, 1], data_new_transposed[:, 2], alpha=0.5)
axs[1, 2].set_xlabel('PC2')
axs[1, 2].set_ylabel('PC3')
axs[1, 2].set_title('Redefined Data (PC2 vs PC3)')

plt.show()
```
